```python
import os
from pathlib import Path
from datetime import datetime

#Librerias necesarias para la aplicacion
!pip install plyer
```

```python


# Cambiar al directorio de trabajo local
os.chdir('/content')

# Inicializar si es necesario
if not Path('buildozer.spec').exists():
    !yes | buildozer init

# Limpiar compilaciones anteriores
!yes | buildozer android clean

# Compilar con respuesta automática "yes" a todoii
!yes | buildozer android debug > build_log.txt 2>&1

!tail -n 50 build_log.txt

# Verificar si se generó el APK
apk_path = Path('/content/bin')
apk_files = list(apk_path.glob('*.apk'))

if apk_files:
    from google.colab import drive
    drive.mount('/content/drive')

    # Crear carpeta en Drive si no existe
    os.makedirs('/content/drive/MyDrive/KIVY', exist_ok=True)

    # Timestamp para el nuevo nombre
    timestamp = datetime.now().strftime('%Y-%m-%d_%H-%M-%S')

    # Copiar con nuevo nombre
    for apk in apk_files:
        new_name = f"{apk.stem}_{timestamp}.apk"
        dest = f"/content/drive/MyDrive/KIVY/{new_name}"
        !cp "{apk}" "{dest}"
        print(f"✅ Copiado: {new_name}")
else:
    print("⚠️ No se encontró ningún archivo .apk en /content/bin/")
```
